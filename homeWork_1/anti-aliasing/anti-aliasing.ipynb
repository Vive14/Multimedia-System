#Anti-aliasing use 

```python
import matplotlib.pyplot as plt
from skimage import io, transform, img_as_float

img = io.imread("barbara.tif", as_gray=True)
img_float = img_as_float(img)

# ไม่ทำ anti-aliasing
downscaled_no_aa = transform.resize(img_float, (img.shape[0] // 2, img.shape[1] // 2), anti_aliasing=False)

# ทำ anti-aliasing (อัตโนมัติ)
downscaled_aa = transform.rescale(img_float, 0.5, anti_aliasing=True)

# ทำ anti-aliasing (กำหนด sigma)
downscaled_aa_sigma1 = transform.rescale(img_float, 0.5, anti_aliasing=True, anti_aliasing_sigma=1)
downscaled_aa_sigma05 = transform.rescale(img_float, 0.5, anti_aliasing=True, anti_aliasing_sigma=0.5)


# แสดงผลเปรียบเทียบ
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].imshow(downscaled_no_aa, cmap='gray')
axes[0, 0].set_title("No Anti-aliasing")
axes[0, 1].imshow(downscaled_aa, cmap='gray')
axes[0, 1].set_title("Anti-aliasing (auto)")
axes[1, 0].imshow(downscaled_aa_sigma05, cmap='gray')
axes[1, 0].set_title("Anti-aliasing (sigma=0.5)")
axes[1, 1].imshow(downscaled_aa_sigma1, cmap='gray')
axes[1, 1].set_title("Anti-aliasing (sigma=1)")

plt.tight_layout()  # ปรับ layout ให้สวยงาม
plt.show()
```
